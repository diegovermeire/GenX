# 1. Refine results within dataframes

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
import seaborn as sns
```

```python


path_cwd = os.getcwd()
full_series_results_folder = "results_19_01_6.43pm_16_epsilon_version"

# outputs_path = joinpath(run_results_folder, string( string("_CO2=", CO2), string("_MinCapReq=",mincap), string("_COMMIT=", commit), string("_epsilon=", eps), string("_V",repetition, "INFEASIBLE")))
CO2Cap = [0, 1, 2]
UCommit = [0, 1]
MinCapReq = [0, 1]
max_repetition = 5
epsilon_values = [0.001 , 0.02559659, 0.04404403, 0.08093892,  0.10553551, 0.12398295, 0.1424304, 0.16087784, 0.19162358, 0.23466761, 0.27771164, 0.31460653, 
0.38224715, 0.57974331, 0.78987166, 1.0]

missing_paths = []
df_missing_folders = pd.DataFrame(columns=['CO2', 'MinCapReq', 'UCommit', 'epsilon', 'repetition'])
df_system_full_settings_results = pd.DataFrame(columns=['CO2', 'MinCapReq', 'UCommit', 'epsilon', 'repetition', 'total_cost', 'Run_time', 'Time_Indexes', 'Build_model_time', 'write_outputs_time', 'termination_Status', 'has_primal'])

for CO2 in CO2Cap:
    for mincap in MinCapReq:
        for commit in UCommit:
            for eps in epsilon_values:
                for repetition in range(1, max_repetition+1, 1):
                    
                    path_result = os.path.join(path_cwd, full_series_results_folder, f"_CO2={CO2}_MinCapReq={mincap}_COMMIT={commit}_epsilon={eps}_V{repetition}")

                    if os.path.exists(path_result):
                        path_result_accuracy =  os.path.join(path_result, "epsilon_values.csv")
                        df_result_accuracy = pd.read_csv(path_result_accuracy)
                        df_system_one_settings_results = pd.DataFrame({
                                        'CO2': CO2,
                                        'MinCapReq': mincap,
                                        'UCommit': commit,
                                        'epsilon': eps,
                                        'repetition': repetition,
                                        'total_cost': df_result_accuracy["total_cost"],
                                        'Run_time': df_result_accuracy["Run_time"],
                                        'Time_Indexes': df_result_accuracy["Time_Indexes"],
                                        'Build_model_time': df_result_accuracy["Build_model_time"],
                                        'write_outputs_time': df_result_accuracy["write_outputs_time"],
                                        'termination_Status': df_result_accuracy["termination_Status"],
                                        'has_primal': df_result_accuracy["has_primal"]
                                    })

                        df_system_full_settings_results = pd.concat([df_system_full_settings_results, df_system_one_settings_results], axis=0)

                    else:
                        print("Missing file with path {}".format(path_result))
                        missing_paths.append(path_result)
                        df_system_one_settings_results = pd.DataFrame({
                                        'CO2': [CO2],
                                        'MinCapReq': [mincap],
                                        'UCommit': [commit],
                                        'epsilon': [eps],
                                        'repetition': [repetition],
                                    })

                        df_missing_folders = pd.concat([df_missing_folders, df_system_one_settings_results], axis=0)
```

# 2. Plot the total cost vs run_time for each GenX setting

```python
# Group by 'CO2', 'MinCapReq', and 'UCommit'
grouped = df_one_system_full_settings_grouped.reset_index().groupby(by =['CO2', 'MinCapReq', 'UCommit'])

# Create subplots
plots_per_row = 2
num_groups = len(grouped)
num_rows = -(-num_groups // plots_per_row)  # Ceiling division

fig, axes = plt.subplots(num_rows, plots_per_row, figsize=(10, num_rows * 5))
axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]

# Iterate through groups and plot
for i, (group_name, group_df) in enumerate(grouped):
    ax = axes[i]
    # Plot all columns in the group
    group_df_ordered = group_df.sort_values(by=["epsilon"], ascending=True)
    ax.plot( group_df_ordered["Run_time"], group_df_ordered["total_cost"], marker='o', linestyle='-', label="{} epsilons".format(len(group_df_ordered)))
    ax.set_xlabel('Run time [s]')
    ax.set_ylabel('Total Cost')
    ax.set_title("CO2cap: {}, MinCapReq: {}, UCommit: {}".format(group_name[0], group_name[1], group_name[2]))
    ax.set_yscale("log")
    # ax.set_xscale("log")
    ax.legend()
    ax.grid()

# Hide unused subplots
for i in range(len(grouped), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

# 3. Plot the accuracy vs. run time for each GenX setting

```python
# Group by 'CO2', 'MinCapReq', and 'UCommit'
grouped = df_one_system_full_settings_grouped.reset_index().groupby(by =['CO2', 'MinCapReq', 'UCommit'])

# Create subplots
plots_per_row = 2
num_groups = len(grouped)
num_rows = -(-num_groups // plots_per_row)  # Ceiling division

fig, axes = plt.subplots(num_rows, plots_per_row, figsize=(10, num_rows * 5))
axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]

# Iterate through groups and plot
for i, (group_name, group_df) in enumerate(grouped):
    ax = axes[i]
    # Plot all columns in the group
    group_df_ordered = group_df.sort_values(by=["epsilon"], ascending=True)
    real_total_cost = group_df_ordered.iloc[0]["total_cost"]
    group_df_ordered["Total Cost Error"] = np.abs(group_df_ordered["total_cost"] - real_total_cost)

    ax.plot( group_df_ordered["Run_time"], group_df_ordered["Total Cost Error"], marker='o', linestyle='-', label="{} epsilons".format(len(group_df_ordered)))
    ax.set_xlabel('Run time [s]')
    ax.set_ylabel('Absolute error on the total cost')
    ax.set_title("CO2cap: {}, MinCapReq: {}, UCommit: {}".format(group_name[0], group_name[1], group_name[2]))
    ax.set_yscale("log")
    ax.grid(True)
    
    # ax.set_xscale("log")
    ax.legend()

# Hide unused subplots
for i in range(len(grouped), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
