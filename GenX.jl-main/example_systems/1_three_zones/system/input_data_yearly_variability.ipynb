```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Fuel resources data 

```python
############################################
#NATURAL GAS PRICES
############################################
cd = os.getcwd()
extra_path = 'Fuels_data.csv'
path_fuels = os.path.join(cd, extra_path)

df = pd.read_csv('Fuels_data.csv')
# Plotting the values for each Time_Index of the 3 columns (CT_NG, ME_NG, MA_NG)
plt.figure(figsize=(10, 6))

plt.plot(df['Time_Index'], df['CT_NG'], label='CT_NG', color='blue')
plt.plot(df['Time_Index'], df['ME_NG'], label='ME_NG', color='green')
plt.plot(df['Time_Index'], df['MA_NG'], label='MA_NG', color='red')

plt.title('Natural Gas Prices Over Time')
plt.xlabel('Time Index')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
