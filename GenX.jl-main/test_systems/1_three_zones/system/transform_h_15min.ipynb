```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Demand_data

```python
cd = os.getcwd()
extra_path_D = 'Demand_data.csv'
path_demand = os.path.join(cd, extra_path_D)
```

```python
df_demand = pd.read_csv(path_demand)
display(df_demand)
```

```python
df_demand_h = df_demand[['Time_Index', 'Demand_MW_z1', 'Demand_MW_z2', 'Demand_MW_z3']].copy(deep=True)
df_demand_h
```

```python
df_demand_15m = df_demand_h.loc[df_demand_h.index.repeat(4)].reset_index(drop=True)
df_demand_15m['Time_Index'] = range(len(df_demand_15m))
df_demand_15m
```

```python
df_demand.drop(df_demand.columns[-4:], axis=1, inplace=True)
df_demand
```

```python
df_demand.loc[0, 'Rep_Periods'] = 1
df_demand.loc[0, 'Timesteps_per_Rep_Period'] = 35040
df_demand.loc[0, 'Sub_Weights'] = 35040
```

```python
df_demand_15m = pd.concat([df_demand,df_demand_15m], axis=1)
display(df_demand_15m)
```

```python
df_demand_15m.to_csv(path_demand, index = False)
```
