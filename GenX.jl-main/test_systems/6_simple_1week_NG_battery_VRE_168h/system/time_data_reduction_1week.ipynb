This script makes sure the orignal homogenous system input data, contains 168 hours. It selects the 168 first hours.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Demand_data.csv

```python
cd = os.getcwd()
extra_path_D = 'Demand_data.csv'
path_demand = os.path.join(cd, extra_path_D)
df_demand = pd.read_csv(path_demand)
display(df_demand)
```

```python
df_demand_h_week = df_demand.iloc[0:168]
display(df_demand_h_week)
```

```python
df_demand_h_week.to_csv(path_demand, index = False)
```

```python
df_demand_h_week.at[0, 'Timesteps_per_Rep_Period'] = 168
df_demand_h_week.at[0, 'Sub_Weights'] = 168
df_demand_h_week.to_csv(path_demand, index = False)
```
